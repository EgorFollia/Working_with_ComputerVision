```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
```

# Image reading

```python
img_1 = cv2.imread("Data/img_1.jpg")
img_2 = cv2.imread("Data/img_2.jpg")
```

```python
plt.figure(figsize=(12, 12))

plt.subplot(1, 2, 1)
plt.imshow(img_1)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_2)
plt.axis('off')

plt.show()
```

# Find the differences

```python
diff = cv2.subtract(img_1, img_2)

gray = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)
ret, mask = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)
diff[mask != 255] = [255, 0, 0] # Paint the blue
```

# Add the mask to the images to vizualize the differences

```python
img_2[mask != 255] = [255, 0, 0]
```

# Vizualization

```python
plt.figure(figsize=(25, 10))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(mask)
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(gray)
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```
